### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
"""
Två sätt att plotta i Matplotlib:
1. Pyplot-tillvägagångssättet: Använder `plt.plot()` och liknande funktioner direkt. Enkelt men mindre flexibelt.
2. Objektorienterat tillvägagångssätt: Skapar `Figure`- och `Axes`-objekt explicit med `plt.subplots()` eller `fig.add_axes()`. Ger mer kontroll.

Rekommenderat tillvägagångssätt: Objektorienterat tillvägagångssätt, eftersom det är mer flexibelt och bättre för komplexa figurer med flera delgrafer.
"""
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
"""
- Figure: Hela "duken" eller fönstret där grafer ritas. Skapas med t.ex. `plt.figure()`.
- Axes: En enskild del av figuren där data plottas (t.ex. en graf eller subplot). Innehåller axlar, titlar, etc.
- Axis: De numeriska axlarna (x-axel, y-axel) i en `Axes`, som definierar skalor och etiketter.
- Artist: Alla objekt som ritas i figuren (t.ex. linjer, text, punkter). `Figure`, `Axes` och `Axis` är typer av artists.
"""
```

### When plotting in Matplotlib, what is the expected input data type?

```python
"""
Förväntad datatyp: NumPy-arrayer eller listor. Matplotlib konverterar automatiskt listor till NumPy-arrayer internt.
"""
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

# Skapa data
x = np.linspace(-4, 4, 100)
y = x**2

# Objektorienterat tillvägagångssätt
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Funktion: y = x^2")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()

# Pyplot-tillvägagångssätt
plt.plot(x, y)
plt.title("Funktion: y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# Skapa figur med två subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Scatterplot
ax1.scatter(random_data_x, random_data_y, alpha=0.5, label="Slumpdata")
ax1.plot(x, y, color='red', label="y = x^2")
ax1.set_title("Scatterplot")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.legend()

# Stapelplot
ax2.bar(names, values, color='green')
ax2.set_title("Fruktdata")
ax2.set_xlabel("Frukt")
ax2.set_ylabel("Antal")

# Justera layout och visa
plt.tight_layout()
plt.show()
```
